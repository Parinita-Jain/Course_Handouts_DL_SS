Imports

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
```

Dense and Sequential object

```python
layer1 = Dense(3, input_dim=2)
layer2 = Dense(2)
layer3 = Dense(1)
```

```python
network = Sequential([layer1, layer2, layer3])
```

```python
network.summary()
```

Neural Network

```python
model = Sequential()
model.add(Dense(3, input_dim=2))
model.add(Dense(2))
model.add(Dense(1))
```

```python
model.summary()
```

input_shape=()

```python
model = Sequential()
model.add(Dense(3, input_shape=(2,)))
model.add(Dense(2))
model.add(Dense(1))
```

```python
model.summary()
```

Input Layer

```python
model = Sequential()
model.add(Input(shape=(2,)))
model.add(Dense(3))
model.add(Dense(2))
model.add(Dense(1))
```

```python
model.summary()
```

Activation function

```python
model = Sequential()
model.add(Dense(3, input_dim=2, activation="tanh"))
model.add(Dense(2, activation="tanh"))
model.add(Dense(1))
```

```python
model.summary()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-100,101)
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
p = sigmoid(x)
```

```python
plt.figure(figsize=(4,4))
plt.plot(p)
plt.show()
```

```python
def tanh(x):
  return (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))
```

```python
op = tanh(x)
```

```python
plt.figure(figsize=(4,4))
plt.plot(op)
plt.show()
```

```python
def relu(x):
  #np.where(condition, true, false)
  return np.where(x>=0, x, 0) 
```

```python
op = relu(x)  
```

```python
plt.figure(figsize=(4,4))
plt.plot(x,op)
plt.show()
```
